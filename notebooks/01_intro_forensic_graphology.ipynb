## What is Forensic Graphology?

**Forensic graphology** (also called *questioned document examination*) is the scientific analysis of handwriting, signatures, and other written material to answer questions relevant to legal proceedings.

Common questions addressed:
- Is this signature genuine or forged?
- Did this person write this anonymous letter?
- Has this document been altered?
- Can we identify personality or emotional state from handwriting?

---

## The Traditional Workflow

A forensic document examiner typically:

1. **Acquires** the questioned document (scan at high resolution, often 600–1200 DPI)
2. **Collects reference samples** from known authors
3. **Visually compares** characteristics: slant, pressure, spacing, letter forms
4. **Writes an expert opinion** for court

**Limitations of manual analysis:**
- Time-consuming on large document sets
- Inherently subjective
- Difficult to reproduce across examiners
- Not scalable for large archives

## Key Concepts

### Handwritten Text Recognition (HTR)
HTR converts handwriting images into machine-readable text. Modern approaches use **sequence-to-sequence Transformers** (like TrOCR) that encode the image with a Vision Transformer and decode it token-by-token with a language model.

### Signature Verification
A **Siamese network** takes two images and learns whether they are from the same class (genuine vs. forged). It does this by mapping both images into an embedding space and measuring their distance — close = same author, far = different/forged.

### Writer Identification
Like fingerprints, handwriting has unique style characteristics. A classifier trained on enough samples can distinguish between writers even on short text fragments.

### Graphological Feature Extraction
Classical computer vision (OpenCV) can objectively measure graphological traits:
- **Slant**: angle of letter strokes relative to vertical
- **Pressure**: mean pixel intensity in stroke regions
- **Spacing**: distribution of whitespace between words and letters
- **Size**: height/width statistics of letter forms

---

## Legal and Ethical Notes

- AI-generated analysis must be reviewed and validated by a qualified forensic document examiner before use in court.
- Model outputs are probabilistic and carry uncertainty — always report confidence scores alongside verdicts.
- Training data biases can affect performance on under-represented handwriting styles, ages, or languages.
- These tools are for **augmenting** expert analysis, not replacing it.
